```python
import omdb
omdb.set_default('apikey', 'f3d1f8f')
movie_name=input()
movie=omdb.search(movie_name)
pd.set_option('display.max_rows',20)
details=pd.DataFrame(movie)
del details['poster']
print(details)
print(details.loc[details.index==0])
```

```python
import requests

from bs4 import BeautifulSoup

import urllib.request

i=details.imdb_id[0]
url='https://www.imdb.com/title/'
rew='/reviews?ref_=tt_urv'
U=url+i+rew
page=requests.post(U)
#print(page.text)
```

```python
comment_page=urllib.request.urlopen(U)
soup = BeautifulSoup(comment_page, 'html.parser')
tbody=soup.find_all('div',attrs={'class':'text show-more__control'})
comment=list()
for element in tbody:
    x=element.text.strip()
    comment.append(x)
print(comment)
```

```python
punctuation='<?''""!,*$&>./\()'
comment_list = [''.join(c for c in s if c not in punctuation) for s in comment]
#print(comment_list)
```

```python
from textblob import TextBlob
for comment in comment_list:
    
    a=TextBlob(comment).sentiment.polarity
print(a)
if -1<=a<-0.5:
    print("not such a good movie")
if -0.5<=a<0:
    print("watchable movie(tolerateable)")
if 0<=a<0.5:
    print("an average movie")
if -1<a<-0.5:
    print("not such a good movie")
```
